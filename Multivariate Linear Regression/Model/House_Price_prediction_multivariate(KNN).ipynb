```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
path = '/gdrive/My Drive/Lemalabs/Assignments/data/'
```

```
data = pd.read_csv(path + 'Chennai_house_price_multivariate_train.csv')
data.head()
```

```
data.isnull().sum(axis = 0)
```

```
data = data.fillna({'N_BEDROOM': data['N_BEDROOM'].mean(),'N_BATHROOM': data['N_BATHROOM'].mean(),'QS_OVERALL': data['QS_OVERALL'].mean()})
```

```
data.isnull().sum(axis = 0)
```

```
x = data.drop(columns=['PRT_ID','DATE_SALE','SALES_PRICE','DATE_BUILD'])
```

```
y = data['SALES_PRICE']
y = np.array(y)
y = y.reshape(-1,1)
x.shape, y.shape
```

```
from sklearn.preprocessing import LabelEncoder

area_encoded = LabelEncoder()
x["AREA"] = area_encoded.fit_transform(x["AREA"])

sale_cond_encoded = LabelEncoder()
x["SALE_COND"] = sale_cond_encoded.fit_transform(x["SALE_COND"])

buildtype_encoded = LabelEncoder()
x["BUILDTYPE"] = buildtype_encoded.fit_transform(x["BUILDTYPE"])

utility_avail_encoded = LabelEncoder()
x["UTILITY_AVAIL"] = utility_avail_encoded.fit_transform(x["UTILITY_AVAIL"])

park_facil_encoded = LabelEncoder()
x["PARK_FACIL"] = park_facil_encoded.fit_transform(x["PARK_FACIL"])

street_encoded = LabelEncoder()
x["STREET"] = street_encoded.fit_transform(x["STREET"])

mzzone_encoded = LabelEncoder()
x["MZZONE"] = mzzone_encoded.fit_transform(x["MZZONE"])
```

```
x.head()
```

```
from sklearn.preprocessing import StandardScaler
x_scaler = StandardScaler()
y_scaler = StandardScaler()

x = x_scaler.fit_transform(x)
y = y_scaler.fit_transform(y)

x.shape, y.shape
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 2)
```

```
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(algorithm = 'brute', n_neighbors=4)

model.fit(x_train, y_train)
```

```
train_pred = model.predict(x_train)
test_pred = model.predict(x_test)

score = model.score(x_train,y_train)
print('train model score : {} %'.format(score*100))

score = model.score(x_test,y_test)
print('test model score : {} %'.format(score*100))
```

```
from sklearn.metrics import mean_absolute_error

train_error = mean_absolute_error(y_train, train_pred)
train_accuracy = 1 - train_error
print('accuracy on train data : {} %'.format(train_accuracy*100))

test_error = mean_absolute_error(y_test,test_pred)
test_accuracy = 1 - test_error
print('accuracy on test data : {} %'.format(test_accuracy*100))
```
